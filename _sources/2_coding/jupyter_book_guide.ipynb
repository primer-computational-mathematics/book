# Jupyter Book guide

This guide provides advice on creating content for ESE Jupyter Book and related etiquette (style). 

## Adding notebooks

In the github directory that creates the jupyter book, only directories with a file 'intro.md' or 'intro.ipynb' will be included in the book. The rest of the directories will be skipped, when creating the website.

```{admonition} Notes on 'intro.md'
* Each directory that includes 'intro.md' will be displayed in the Table of Contents with by the first header from the 'intro.md' file. For header style guide, go to {ref}`this section <style_guide_headers>`. Always check if your headers are consistent with other on the website!
* Do not add Markdown tables of contents into your notebooks, the website will create one in top right corner.
```

## File naming

Currently, the files are sorted with a [natural sorting algorithm](https://rosettacode.org/wiki/Natural_sorting), meaning that files are not only sorted in alphabetical order, but also in a human order if numbers appear in the file name. This way, if you want your files to appear in the Table of Contents in a specified order, you can add integers at the begining of each file to sort them accordingly:

```
'1_basics.ipynb'
'2_sorting.ipynb'
...
'11_merging.ipynb'
'12_grouping.ipynb'
```

```{admonition} Notes on file naming style
* Make your file names consistent and descriptive
* Use underscores instead of spaces in names
* Do not use brackets in your file names
```

## Headers in notebooks

To make headers in Markdown cell you can use \#:
```
# Header level 1
## Header level 2
### Header level 3
#### Header level 4
##### Header level 5
###### Header level 6
```
The first section title with a \# in your notebook will produce the title that displays in the Table of Contents on the website. The name of the file is used for sorting purposes in the Table of Contents.

```{admonition} Notes on header style
* Start your notebook with a single \# header and do not make another one.
* Please capitilise first letter of each header and leave the rest lowercase, unless it is a proper name.
```

## Cross-referencing notebooks/sections
Sometimes it is useful to reference something from other notebook. For example, we would like to remind the user where to find a guide on writing maths in jupyter notebooks. The guide can be accessed from {ref}`style_guide_maths` and the link takes us to the appropriate section. To make sure we can reference a section in a notebook, we need to label them first:

```
(this_is_a_label)=
# This is a section title that we want a label for
```
and then in text, you reference it as

```
This is some text and I want to reference my new label, {ref}`this_is_a_label` and a label is referenced
```

This method works for all labels within your jupyter book.

You can also use your own text instead of the name of the referenced section, e.g. you can find math style guide {ref}`here <style_guide_maths>`. This can be produced with syntax

    Some text here and then you can find it {ref}`here <this_is_a_label>`.

```{admonition} You can use cross-referencing to create customised tables of contents:
:class: dropdown, tip

1. {ref}`printing_intro`
    - {ref}`basic_printing`
    - {ref}`formatted_printing`
    - {ref}`printing_exercises`
2. {ref}`variables`
    - {ref}`variable_types`
    - {ref}`variable_operations`
3. {ref}`if_statement`
4. {ref}`while_loops`
```

```{admonition} Notes on cross-referencing style
* Make labels descriptive.
* Do not make simple labels in your notebooks, e.g. "introduction", "basics". This way we will avoid double referencing and pointing to wrong notebooks.
* The official website provides extra syntax for referencing documents:
    
    {doc}`path/to/doc'
    
Be careful when using this style, as files may move/change names. Better to keep a label for the main header in the notebook, as it is more to remain unchanged.
    
```

## Maths and equations

### In-line equations

If you would like to include in-line equation in the Markdown text, e.g. to explain that area \\(A=\pi r^2\\), you should follow this syntax:

```
Your text goes here \\(A = \pi r^2\\) and more text here.
```

Equations use standard LaTeX syntax. For some examples how to write equations in LaTeX, look [here](https://en.wikibooks.org/wiki/LaTeX/Mathematics). For a comprehensive list with math symbols look [here](https://oeis.org/wiki/List_of_LaTeX_mathematical_symbols).


### Math blocks
If you would like your equation to display in the next line and centred, e.g. the area of a circle is

\\[A=\pi r^2,\\]

follow this syntax:

```
Your text goes here and then:

\\[A=\pi r^2,\\]

more of your text goes here.
```

```{admonition} Notes on equation style
* Please use appropriate punctuation around equations, do not forget to put a full stop when equation ends a sentence.
* Break down long equations into more lines.
* Do not copy special characters from other websites, find a LaTeX symbol that can be used to create such a character.
```

## Hiding and removing code cells
If you want to hide or remove some parts of the notebook from the website, you can add tags to code cells. To enable tags in a jupyter notebook, go to View -> Cell Toolbar -> Tags. In the right hand corner of each cell you will be able to enter a tag.

```{admonition} Tag types
:class: tip, dropdown
* **hide-input** tag - hides the code but the user can toggle it to see
* **hide-output** tag - hides the output but the user can toggle it to see
* **hide-cell** tag - hides code and output but the user can toggle it to see
* **remove-input** tag - will remove the code but the output will stay on display
* **remove-output** tag - will keep the code but the remove the output from display
* **remove-cell** tag - will remove both the code and output and user won't see it on the website
```

```{admonition} Notes on hiding and removing cells
* It is best to hide cells with code for plotting unless it is crucial to explain the type of plot used
* Hide cells with importing libraries 
```

## Special content blocks

[Jupyter Book website](https://jupyterbook.org/reference/cheatsheet.html#admonitions) provides an excellent guide for creating special content blocks.

Most commonly use special content block is a blue admonition box is:

```{admonition} Title
text
```
which can be created with this syntax:
````
```{admonition} Title
text
```
````

You can hide your box, using following syntax:

````
```{admonition} Title
:class: dropdown
text
```
````
which will display as:
```{admonition} Title
:class: dropdown
text
```

## Adding index
To attach an index to a section in your notebook, use syntax:
````
```{index} index name
```
````

```{admonition} Notes on index style
* Make index with capital letter first
* If you are referring to e.g. merging in Pandas, make index "Pandas: merging"
* Do not make overly complicated index names
```

## Figures

### Figures from github repository (does not work locally!)
To add figure from github repository, use following syntax:

````
```{figure} path/to/figure
:width: 100px
```
````

```{figure} ../images/logo/logo.png
:width: 100px
```

See [Jupyter Book website](https://jupyterbook.org/content/figures.html) for more options when creating figures.

### Figures from websites

Use following syntax for figures found on websites:

```
<img src="https://upload.wikimedia.org/wikipedia/commons/d/d4/Gladiator_frog_Osa.JPG" width="200px">
```

### Margin content

Sometimes you want to add extra content that is useful, but it does not fit in the main text. You can use right margin for that with syntax:

````
``` {margin} Title
Some text in the margin
```
````

### Links to websites
If you would like some text to refer to an online website follow this syntax:

| Syntax    | Example    | Result  |
| :-------- | ---------- | ------: |
| ```[text](target)``` | ```[ESE at Imperial](https://www.imperial.ac.uk/earth-science/)```          | [ESE at Imperial](https://www.imperial.ac.uk/earth-science/)       |
| ```<target>``` | ```<https://www.imperial.ac.uk/earth-science/>``` | <https://www.imperial.ac.uk/earth-science/>|

```{admonition} Notes on link style
* Use the top option for displaying links, unless it is very crucial to keep the website address intact.
```
