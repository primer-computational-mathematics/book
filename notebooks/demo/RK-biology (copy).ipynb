### Exercise: Population Growth

Given the population $P$ of a species in a closed ecosystem, we can model the population growth as a function of the reproductive capabilities of the species and the "carrying capacity" or the maximum feasible population relative to the resources available in the ecosystem. As such, we can model the system using the logistic model growth formula

$$ \frac{dP}{dt} = kP(M-P) $$

where $M$ is the carrying capacity, $P$ is the population at any given time, and $k$ is a constant representing the rate of reproduction of the species. Using a value of $k = 0.2$ and $M = 1000$, model the population growth of species with initial populations $P_0 = 20$ and $P_0 = 3000$ over a time frame of $30$ years, using time-steps of 1 day at a time.
